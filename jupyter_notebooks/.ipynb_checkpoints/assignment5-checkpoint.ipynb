```python
import numpy as np
import sys

A = np.matrix(np.load('assignment0828_A.npy'))
b = np.matrix(np.load('assignment0828_b.npy'))

x = np.matrix([0] * 2000).transpose()
k = 5000
ro = 1
t = 1/(max((np.linalg.eig(A.transpose() * A))[0]).real)

def T(x):
    res = np.matrix(
        [[(1 if x.item(i,0)>=0 else -1)*np.amax(np.absolute(x.item(i,0))-ro*t,0)]
         for i in range(2000)])
    return res

for i in range(k):
    prev = A*x-b
    if i%50 == 0:
        sys.stdout.write('\r\033[94m' + '\u25A0'*(int(i/50)+1) 
                         + '\033[0m'  + int(i/50+1).__str__() + "%")
        sys.stdout.flush()
    x = T(x - t*A.transpose()*(prev))
sys.stdout.write('\r')
print(A*x - b)
```
