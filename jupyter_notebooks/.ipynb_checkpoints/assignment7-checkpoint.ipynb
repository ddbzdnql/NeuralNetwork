```python
import numpy as np
import analytic_function as af
import backprop as bp
from backprop import Node as node

x = node('x')
#x.value = np.load('assignment7_X.npy')
w1 = node('w1')
#w1.value = np.load('assignment7_W1.npy')
w2 = node('w2')
#w2.value = np.load('assignment7_W2.npy')

h1 = node('h1')
h2 = node('h2')

x.toNode(h1)
w1.toNode(h1)
h1.toNode(h2)
w2.toNode(h2)

h1.analyze(af.funcMatVec(), [w1, x], True)
h2.analyze(af.funcMatVec(), [w2, h1], True)

x_val = np.load('assignment7_X.npy')
w1_val = np.load('assignment7_W1.npy')
w2_val = np.load('assignment7_W2.npy')

bp.forward([x, w1, w2], [x_val, w1_val, w2_val])

bp.backward(h2)
print("gradient of W1")
print("[")
for i in w1.der:
    print("[")
    for j in i:
        print("[", end = "\0")
        for k in j:
            print("%5s," % (k.__str__()), end = "\0")
        print("]")
    print("]")
print("]")

print("\ngradient of W2")
print("[")
for i in w2.der:
    print("[")
    for j in i:
        print("[", end = "\0")
        for k in j:
            print("%6s," % (k.__str__()), end = "\0")
        print("]")
    print("]")
print("]")

h = 0.01
w1h = [[j + h for j in i] for i in w1.value]
v = np.dot(w1h, x.value)
difference = [v[i] - h1.value[i] for i in range(len(v))]
```
